```python
#final query for this data analysis:

query = """
        SELECT subscriber_id, 
               call_date,
               block_audio_percent_listened,
               km_block_reached,
               km_block_reached,
               listen_menu_or_digest
               listener, 
               gender, 
               age,
               organization_country, 
               lower(Location) as location, 
               location_level_2 as state, 
               duration_listened_minutes, 
               income_source, 
               highest_expense_area, phone_type, 
               subscriber_pereferred_language, block_title
        FROM `viamo-datakind.datadive.321_sessions_1122`
        WHERE gender != 'null' AND education_level = 'secondary_school'
        """
Bigquery_client = bigquery.Client()
df = pd.read_gbq(query)
   
display(df)

#save as CSV

df.to_csv('C:/Users/Chandra/Desktop/VIAMO-py/viamo_SecSchool_Nigeria_subset_dataV3.csv')
```
